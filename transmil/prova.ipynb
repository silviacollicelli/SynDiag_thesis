```python
from torchmil.datasets import BinaryClassificationDataset
from torchmil.models import TransMIL
from data_trans import transmil_collate, make_deterministic_dataloader
import torch
from torch.utils.data import DataLoader
```

```python
dataset = BinaryClassificationDataset(features_path, labels_path, bag_keys=["X", "Y"], verbose=False, load_at_init=False)
```

```python
def transmil_collate_bs1(batch):
    # batch is a list of length 1
    item = batch[0]
    X = item["X"].unsqueeze(0)  # (1, bag_size, 1024)
    Y = item["Y"].unsqueeze(0)  # (1,)
    return {"X": X, "Y": Y}
```

```python
dataloader = DataLoader(
        dataset=dataset,
        batch_size=1,
        shuffle=True,
        collate_fn=transmil_collate_bs1
    )
```

```python
model = TransMIL((1024,)).to(device)
```

```python
for batch in dataloader:
    print(batch['X'].size())
    X = batch["X"].to(device)  # (B, bag_size, 1024)
    Y = batch["Y"].to(device)
    print(Y)
    pred = model(X)
    print(pred)
    break
    #loss = model.criterion(pred, Y.float())
```
